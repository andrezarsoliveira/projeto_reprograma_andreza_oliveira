```python
df = pd.read_csv("/running_perfomances.csv", sep="," , encoding= 'UTF-8')

paises_unicos = df['Country'].nunique()

print(f"O número de países únicos na tabela é: {paises_unicos}")
```

```python
contagem_atletas_por_evento = df['Event'].value_counts().head(5)

plt.figure(figsize=(12, 6))
contagem_atletas_por_evento.plot(kind='bar', color= '#9A32CD')
plt.title("As 5 Provas mais populares no Atletismo")
plt.xlabel("Tipo de Prova")
plt.ylabel("Número de Atletas")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
cidades = df['City'].nunique()

print(f"O número de cidades na tabela é: {cidades}")
```

```python
df = pd.read_csv("/running_perfomances.csv", sep="," , encoding= 'UTF-8')
contagem_atletas_por_evento = df['City'].value_counts().head(10)

plt.figure(figsize=(12, 6))
contagem_atletas_por_evento.plot(kind='bar')
plt.title("As 10 Cidades mais populares no Atletismo")
plt.xlabel("Cidades")
plt.ylabel("Cidades que receberam mais vezes os eventos")
plt.xticks(rotation=45, ha="right")
plt.show()
```
